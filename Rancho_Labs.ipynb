```
import gradio as gr

def calculatorv1(number1, number2):
  sum1= number1+number2
  sub1= number1-number2
  mul1 = number1*number2
  div1= number1/number2
  return sum1, sub1, mul1, div1

app= gr.Interface(
    fn= calculatorv1,
    inputs=[gr.Number("Enter first number:"), gr.Number("Enter second number:")],
    outputs= [gr.Number(label="Addittion"),
              gr.Number(label="SuBratcion"),
              gr.Number(label="Multiplication"),
              gr.Number(label="Divivsion")],
    title="Simple Calculator"
)

app.launch()
```

```
def HCF(au, av):
  if (au==0 or av==0):
    return "HCF is not defined."
  while (av !=0):
    aw=au%av
    au=av
    av=aw
  return au
ax = int(input("Enter first number:"))
ay = int(input("Enter second number:"))
print(f"The HCF of {ax} and {ay} is ({HCF(ax, ay)}).")
```

```
import gradio as gr

def HCF(au, av):
  if (au==0 or av==0):
    return "HCF is not defined."
  while (av !=0):
    aw=au%av
    au=av
    av=aw
  return au

app= gr.Interface(
    inputs=[gr.Number(label="Enter first number"),
            gr.Number(label="Enter second number")],
    fn=HCF,
    outputs=[gr.Textbox(label="The HCF is:")]
)

app.launch()
```

```
import gradio as gr
def factorial(number):
  finalNumber = 1
  while (number != 0):
    finalNumber = finalNumber * number
    number = number -1
  return finalNumber
app = gr.Interface(
    inputs= gr.Number(label="Enter number to find factorial of:"),
    fn = factorial,
    outputs = gr.Number(label="Factorial")

)
app.launch()
```
