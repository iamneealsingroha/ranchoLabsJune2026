```python
import gradio as gr
def multiply(num1, num2):
  finalAnswer = []
  for answer in range(1, num2+1):
    resutlt= num1 * answer
    finalAnswer.append(f"{num1} x {answer} = {resutlt}")
  return finalAnswer
app = gr.Interface(
    inputs = [gr.Number(label="Enter number to multiply:"),
              gr.Number(label="Enter number to multiply till:")],
    fn = multiply,
    outputs = [gr.Textbox(label="Answer:")]
)
app.launch()
```

```python
neighbours={
    "A":{"C":3, "F":2},
    "B":{"D":1, "E":2, "F":6, "G":2},
    "C":{"A":3, "D":4, "E":1, "F":2},
    "D":{"B":1,"C":4},
    "E":{"B":2,"C":1, "F":3},
    "F":{"A":2, "B":6, "C":2, "E":3, "G":5},
    "G":{"B":2, "F":5}
}
def find_path(start_city,target_city):
  global neighbours
  shortestPath ={}
  unexplored = []
  for city in neighbours:
    shortestPath[city] = float('inf')
    unexplored.append(city)
  shortestPath[start_city] = 0
  unexplored.remove(start_city)
  current_city = start_city
  while current_city != target_city:
    for next_city in neighbours[current_city]:
      newval = shortestPath[current_city]+ neighbours[current_city][next_city]
      if newval < shortestPath[next_city]:
        shortestPath[next_city] = newval
    next_closest_city = unexplored[0]
    for city in unexplored:
      if shortestPath[city] < shortestPath[next_closest_city]:
        next_closest_city = city
    print("next closest city= ", next_closest_city)
    unexplored.remove(next_closest_city)
    current_city = next_closest_city
  return f"Shortest Path = {shortestPath[target_city]}"
import gradio as gr
app = gr.Interface(
    fn = find_path,
    inputs =[
        gr.Textbox(label="Start City"),
        gr.Textbox(label="Target City"),
    ],
    outputs = gr.Textbox(label="Result")
)
app.launch(debug=True)
```

```python
import gradio as gr

# Responses dictionary
responses = {
    "i feel sad": "I'm sorry to hear that. What's bothering you?",
    "i am lonely": "Feeling lonely can be tough.",
    "i'm stressed": "What's been stressing you out?",
    "i feel lost": "It's okay to feel lost sometimes.",
    "i need someone to talk to": "Tell me what's on your mind, and I'll listen.",
    "i feel worthless": "Can you share more about what's making you feel this way?",
    "i can't sleep": "Have you tried deep breathing or listening to soothing music before bed?",
    "i feel hopeless": "Would you like to talk about what's causing these feelings?",
    "i feel angry": "Anger can be intense. Do you want to talk about what's making you angry?",
    "i need motivation": "Starting small and celebrating small wins can make a big difference.",
    "i feel empty": "Sometimes, connecting with loved ones or engaging in hobbies can help.",
    "i'm nervous": "Try focusing on your breathing or thinking about something that makes you happy.",
    "i don't know what to do": "Let's break it down together. What's been on your mind?"
}

# Function to handle chat responses
def chatbot(question, chat_history):
    if chat_history is None:
        chat_history = []

    bot_response = "Sorry, I didn't understand that."

    for key in responses:
        if question.lower().split() == key.split():
            bot_response = responses[key]
            break

    if question.lower().split() == ["bye"]:
        bot_response = "Okay, bye!"

    # Store conversation
    chat_history.append((question, bot_response))

    return chat_history, chat_history


print("Hello, how can I assist you?")

# Gradio Interface
app = gr.Interface(
    fn=chatbot,
    inputs=[
        gr.Textbox(
            label="How may I help you?",
            placeholder="Type here..."
        ),
        gr.State([])
    ],
    outputs=[
        gr.Chatbot(label="Chat History"),
        gr.State([])
    ],
    title="Mental Health Chatbot"
)

# Launch the chatbot
app.launch()
```
